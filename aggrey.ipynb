```python
# reading the data
import pandas as pd
kc_house_data = pd.read_csv("kc_house_data.csv")
kc_house_data
```

```python
# simplifying the data
kc_house_data_simple = kc_house_data.loc[:,["id","price","bedrooms", "bathrooms", "yr_built"]]
kc_house_data_simple
```

```python
# checking for null values
kc_house_data_simple.isnull().sum()
```

```python
# there are no null values
```

```python
#checking for duplicates
kc_house_data_simple.duplicated().sum()
```

```python
# there are three duplicates
# droping the duplicates
kc_house_data_simple.drop_duplicates()
```

```python
# Checking the correlation
# Calculate correlation matrix
corr_matrix = kc_house_data_simple.corr()

print(corr_matrix)
```

```python
# Visualize correlation matrix as a heatmap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=.5,  vmin=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# The strongest correlation observed is between bathrooms and price at 0.526 indicating that the number of bathrooms has a relatively stronger influence on the price compared to other variables in the matrix.
# Interpreting weak correlations: For variables like id and yr_built, where correlations are close to zero (-0.017 to 0.054), it suggests that these factors have little to no linear relationship with the other variables in the dataset
```

```python
#plotting pairplots to view the correlations
sns.pairplot(kc_house_data_simple)
```

```python
# modelling while considering the price and bathrooms
import statsmodels.api as sm
# for independent variable
X = kc_house_data_simple['bathrooms']
#for dependent variable
y = kc_house_data_simple['price']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
r_squared = model.rsquared
print(f"R-squared: {r_squared}")
print(model.summary())
```

```python
#   R squared value is approxiately 27.658% indicating other factors affect the price
```

```python
# modelling while considering the price, bedrooms
import statsmodels.api as sm
# for independent variable
X = kc_house_data_simple['bedrooms']
#for dependent variable
y = kc_house_data_simple['price']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
r_squared = model.rsquared
print(f"R-squared: {r_squared}")
print(model.summary())
```

```python
#   R squared value is approxiately 9.5% indicating other factors affect the price
```

```python
# modelling while considering the price, yr_built
import statsmodels.api as sm
# for independent variable
X = kc_house_data_simple['yr_built']
#for dependent variable
y = kc_house_data_simple['price']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
r_squared = model.rsquared
print(f"R-squared: {r_squared}")
print(model.summary())
```

```python
#   R squared value is approxiately 0.3 % indicating other factors affect the price
```
